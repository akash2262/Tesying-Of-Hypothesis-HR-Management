Mean Comparison

H0: Assumed value

H1: Assumed value is not close to population stats

```python
from scipy import stats

null_hypothesis_value = 3.5

# Perform t-test
t_statistic, p_value = stats.ttest_1samp(df['Rating'], null_hypothesis_value)

# Output results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Check if the result is statistically significant (common significance level 0.05)
if p_value < 0.05:
    print("The result is statistically significant - reject the null hypothesis.")
else:
    print("Fail to  reject the null hypothesis.")
```

Testing the differance;

H0= Mean_Rating(male)-Mean_Rating(female)=0

H1: Non Zero

```python
groups = df.groupby('Gender')['Rating']
t_statistic, p_value = stats.ttest_ind(*[group.values for name, group in groups])

# Output results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Check if the result is statistically significant (common significance level 0.05)
if p_value < 0.05:
    print("The result is statistically significant - reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
groups = df.groupby('Profile')['Rating']
t_statistic, p_value = stats.ttest_ind(*[group.values for name, group in groups])

# Output results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Check if the result is statistically significant (common significance level 0.05)
if p_value < 0.05:
    print("The result is statistically significant - reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

Increase in rating with experiennce

```python
import statsmodels.api as sm

# Add a constant term to the independent variable
X = sm.add_constant(df['Experience'])
y = df['Rating']

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())
```

Coeffiencent of experience is not significant.

```python
X = sm.add_constant(df[['Experience', 'Profile']])
y = df['Rating']

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())
```

P value is improved .
